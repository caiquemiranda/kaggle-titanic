# Tratamento e uso de Machine Learning

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
modelo = RandomForestClassifier(n_estimators=100, n_jobs=1, random_state=0)
```

```python
train['Sex'].value_counts()
```

```python
def sex_b(x):
    if x == 'female':
        return 1
    else:
        return 0 
```

```python
train['Sex_b'] = train['Sex'].map(sex_b)
```

```python
# variaveis usadas para treinar modelo
v = ['Sex_b', 'Pclass','Parch','SibSp','Age']
```

```python
train.head()
```

```python
X = train[v] # variaveis usadas 
y = train['Survived'] # objetivo de previsão
```

```python
# substituição dosvalores faltantes 
X = X.fillna(-1)
```

```python
modelo.fit(X, y)
```

```python
test['Sex_b'] = test['Sex'].map(sex_b)
```

```python
X_previsoes = test[v]
X_previsoes.head()
```

```python
X_previsoes = X_previsoes.fillna(-1)
```

```python
prev = modelo.predict(X_previsoes)
```

```python
prev
```

```python
test.head()
```

```python
submit = pd.Series(prev, index=test['PassengerId'], name='Survived')
submit.shape
```

```python
submit.to_csv('fist_model_ot2.csv', header=True)
```
